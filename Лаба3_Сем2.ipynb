```python
def f(y, t):
  try:
    return (1/(t ** 2) * math.cos(y))
  except ZeroDivisionError:
    print('ZeroDivisionError')
y0 = 1
t = np.linspace(0.1, 1, 10)
y = odeint(f, y0, t)
plt.plot(t, y)
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
